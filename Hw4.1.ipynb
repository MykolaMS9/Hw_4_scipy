```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
speed = np.array([25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65])
speed
```

```python
time = np.linspace(0,11, 12)
time
```

Виконайте виведення графіка — значення точок швидкості (plot або scatter). Задайте розмір області, що відображається (0, 11) і (0, 130). Задайте відображення сітки

```python
plt.plot(time, speed, label="виміряна швидкість")
plt.grid()
plt.xlabel("Проміжки виміру, год", fontsize="small", color="midnightblue")
plt.ylabel("Швидкість, км/год", fontsize="small", color="midnightblue")
plt.title("Графік швидкості від часу", fontsize=15)
plt.legend()
plt.xlim(0, 11)
plt.ylim(0, 130)
```

Виконайте інтерполяцію за допомогою interp1d(kind='cubic') та отримайте функцію на 10000 значень. Побудуйте безперервний графік отриманої функції.

```python
data_inter = sp.interpolate.interp1d(time, speed, kind='cubic')
data_inter
```

```python
time_cubic = np.linspace(0, 11, 10000)
plt.plot(time_cubic, data_inter(time_cubic), label="кубічна інтерполяція швидкості")
plt.plot(time, speed, label="виміряна швидкість")
plt.grid()
plt.xlabel("Проміжки виміру, год", fontsize="small", color="midnightblue")
plt.ylabel("Швидкість, км/год", fontsize="small", color="midnightblue")
plt.title("Графік швидкості від часу", fontsize=15)
plt.legend()
plt.xlim(0, 11)
plt.ylim(0, 130)
```

Обчисліть інтеграл для отриманої інтерполяційної функції на проміжку [0, 11]

```python
sp.integrate.quad(data_inter, 0, 11)
```

Виконайте також пункти 4 та 5 для kind='quadratic'

```python
data_quadratic = sp.interpolate.interp1d(time, speed, kind='quadratic')
data_quadratic
```

```python
time_cubic = np.linspace(0, 11, 10000)
plt.plot(time_cubic, data_quadratic(time_cubic), label="квадратична інтерполяція швидкості")
plt.plot(time_cubic, data_inter(time_cubic), label="кубічна інтерполяція швидкості")
plt.plot(time, speed, label="виміряна швидкість")
plt.grid()
plt.xlabel("Проміжки виміру, год", fontsize="small", color="midnightblue")
plt.ylabel("Швидкість, км/год", fontsize="small", color="midnightblue")
plt.title("Графік швидкості від часу", fontsize=15)
plt.legend()
plt.xlim(0, 11)
plt.ylim(0, 130)
```
